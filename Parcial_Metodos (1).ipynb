# Modelos de Clasificación

**Pregunta Teórica**

En este caso, el objetivo es predecir si un prestamo sera aprobado o rechazado en funcion de las caracterısticas del solicitante. ¿Que tipo de modelo de clasificacion usarıas y por que? Discute las ventajas y desventajas de este modelo

**Respuesta:**
Para predecir si un préstamo será aprobado o rechazado en función de las características del solicitante, un modelo adecuado sería el árbol de decisión. Este tipo de modelo es fácil de interpretar, puede manejar tanto variables numéricas como categóricas sin necesidad de escalarlas, y capta relaciones no lineales entre las variables, lo cual es útil en contextos donde las decisiones pueden depender de múltiples combinaciones de factores.

Ventajas:
- Fácil de interpretar y visualizar.
- No requiere escalado de variables.
- Puede manejar tanto variables numéricas como categóricas.
- Captura relaciones no lineales entre variables.
- Resistente a valores faltantes y outliers.
- Útil para conocer qué variables son más importantes en la decisión.

Desventajas:
- Propenso al overfitting si no se controla la profundidad del árbol.
- Sensible a pequeños cambios en los datos (puede cambiar mucho la estructura).
- Puede generar modelos demasiado complejos si no se poda.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE
```

```python
Distribución=data[' loan_status'].value_counts(normalize=True)
Distribución
```

```python
import matplotlib.pyplot as plt

Distribución.plot(kind='bar')
plt.title('Class Distribution of loan_status')
plt.xlabel('Loan Status')
plt.ylabel('Frequency')
plt.show()
```

```python
x = data[[' income_annum',' loan_amount',' residential_assets_value',' commercial_assets_value',' luxury_assets_value',' bank_asset_value']].values
y = data[' loan_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y, random_state=42)
print("Dimensiones del conjunto de entrenamiento:", X_train.shape)
print("Dimensiones del conjunto de prueba:", X_test.shape)
```

```python
if y_train.value_counts(normalize=True).max() > 0.60:
    print("Aplicando SMOTE para balancear el conjunto de entrenamiento...")
    smote = SMOTE(random_state=42)
    X_train, y_train = smote.fit_resample(X_train, y_train)
    print("Nueva distribución de clases:")
    print(pd.Series(y_train).value_counts(normalize=True))
```

```python
#Árbol de Decisión
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
#Regresión Logística
lr_model = LogisticRegression(max_iter=10000, random_state=42)
lr_model.fit(X_train, y_train)
```

```python
y_pred_dt = dt_model.predict(X_test)
y_pred_lr = lr_model.predict(X_test)
```

```python
print("Reporte de Clasificación - Árbol de Decisión (Test):")
print(classification_report(y_test, y_pred_dt))
```

```python
print("Reporte de Clasificación - Regresión Logística (Test):")
print(classification_report(y_test, y_pred_lr))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))
sns.heatmap(confusion_matrix(y_test, y_pred_dt), annot=True, fmt='d', ax=ax[0], cmap='Greens')
ax[0].set_title("Matriz de Confusión - Árbol de Decisión (Test)")
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt='d', ax=ax[1], cmap='Blues')
ax[1].set_title("Matriz de Confusión - Regresión Logística (Test)")
plt.show()
```

```python
y_pred_train_dt = dt_model.predict(X_train)
y_pred_train_lr = lr_model.predict(X_train)
```

```python
print("Reporte de Clasificación - Árbol de Decisión (Train):")
print(classification_report(y_train, y_pred_train_dt))
```

```python
print("Reporte de Clasificación - Regresión Logística (Train):")
print(classification_report(y_train, y_pred_train_lr))
```

```python
y_prob_dt = dt_model.predict_proba(X_test)[:, 1]
y_prob_lr = lr_model.predict_proba(X_test)[:, 1]
```

```python
y_test_numeric = y_test.map({' Approved': 1, ' Rejected': 0})
```

```python
from sklearn.metrics import roc_curve, auc

fpr_dt, tpr_dt, _ = roc_curve(y_test_numeric, y_prob_dt)
fpr_lr, tpr_lr, _ = roc_curve(y_test_numeric, y_prob_lr)
auc_dt = auc(fpr_dt, tpr_dt)
auc_lr = auc(fpr_lr, tpr_lr)
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr_dt, tpr_dt, label=f"Árbol de Decisión (AUC = {auc_dt:.2f})")
plt.plot(fpr_lr, tpr_lr, label=f"Regresión Logística (AUC = {auc_lr:.2f})")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.title("Curva ROC")
plt.legend()
plt.show()
```
